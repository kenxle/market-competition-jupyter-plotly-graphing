```python
# Read in file
import numpy as np
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

end_period=60
start_period=32
sdir=str(end_period)+'/'
filename='w02_MBA_c1_out.xlsm'


for cur_period in range(start_period, end_period+1):
    sheetname='P'+str(cur_period)
    print('file: ' + sdir+filename + ' | sheetname: ' + sheetname)
    df = pd.read_excel(sdir+filename, sheet_name=sheetname, header=1)
    # print(df[0])
    # Name the first column
    cols = list(df.columns.values)
    cols[0] = "Attribute"
    df.columns = cols
    df['period'] = cur_period
    
    if cur_period == start_period:
#         print('set initial df')
        df_public_countries = df.head(49).copy()
    else:
#         print('append df')
        df_public_countries = df_public_countries.append(df.head(49).copy())
        
#     display(df_public_countries)
        
df_public_countries = df_public_countries.drop(['Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], axis=1)


# print(df_public_countries.loc['Demand - Units P1']['JAPAN'])
display(df_public_countries.head())
    # dflat_public_countries = df_public_countries.unstack(level=0)
    # print(dflat_public_countries.head())
```

```python
def label_dim (row):
   if row['Price'] != 1 :
      return 'Price'
   if row['Quality'] != 1 :
      return 'Quality'
   if row['Marketing'] != 1:
      return 'Marketing'
   if row['Green']  != 1:
      return 'Green'
df['dimension'] = df.apply (lambda row: label_dim (row),axis=1)
```
